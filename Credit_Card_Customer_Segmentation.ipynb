```python
col = list(df.columns)
for i in col:
    if df[i].dtypes == 'int64' or df[i].dtypes == 'float64':
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.ylabel('count')
        plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['CUST_ID'] = le.fit_transform(df['CUST_ID'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scaler = StandardScaler()
df_arr = scaler.fit_transform(df)
data_scaled = pd.DataFrame(df_arr,columns = df.columns)
data_scaled
```

```python
df = df.drop(['CUST_ID'],axis=1)
df
```

```python
data_scaled.describe().T
```

```python
ss = {}

for i in range(1,10):
    k = KMeans(n_clusters = i,random_state = 1).fit(data_scaled)
    ss[i] = k.inertia_

ss
```

```python
plt.plot(list(ss.keys()),list(ss.values()),'rx--')
```

```python
#WCSS(within cluster sum of squares)
wcss = {}
for i in range(2,10):
    k = KMeans(n_clusters=i,random_state=1).fit(data_scaled)
    labels = k.predict(data_scaled)
    wcss[i] = silhouette_score(data_scaled,labels)

wcss
```

```python
plt.plot(list(wcss.keys()),list(wcss.values()),'gx--')
```

```python
k = KMeans(n_clusters=5,random_state=1).fit(data_scaled)
df['KMean_Labels'] = k.predict(data_scaled)
df.head()
```

```python
df['KMean_Labels'].value_counts()
```

```python
df['KMean_Labels'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='BALANCE',y='PURCHASES',hue='KMean_Labels',palette='viridis')

plt.title('K-Means Clustering Results')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.legend(title='Clusters')
plt.show()
```
